```python
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
df[['Attrition', 'AttritionFlag']].head()
```

```python
dept_attrition_rate = (
    df.groupby('Department')['AttritionFlag']
    .mean()
    .sort_values(ascending=False)
) * 100

dept_attrition_rate
```

```python
overall_attrition_rate = df['AttritionFlag'].mean() * 100
overall_attrition_rate
```

```python
plt.figure(figsize=(8,5))
dept_attrition_rate.plot(kind='bar')
plt.title("Attrition Rate (%) by Department")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

Departments like Sales and Human Resources experience higher attrition compared to R&D.

```python
df['SalaryBand'] = pd.qcut(
    df['MonthlyIncome'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
salary_attrition_rate = (
    df.groupby('SalaryBand', observed=True)['AttritionFlag']
    .mean()
) * 100
```

```python
plt.figure(figsize=(8,5))
salary_attrition_rate.plot(kind='bar')
plt.title("Attrition Rate (%) by Salary Band")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

Attrition is highest among employees in lower salary bands, indicating pay as a strong driver of employee turnover.

```python
job_attrition_rate = (
    df.groupby('JobSatisfaction')['AttritionFlag']
    .mean()
) * 100
```

```python
plt.figure(figsize=(8,5))
job_attrition_rate.plot(kind='bar')
plt.title("Attrition Rate (%) by Job Satisfaction")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

```python
overtime_attrition_rate = (
    df.groupby('OverTime')['AttritionFlag']
    .mean()
) * 100
```

```python
plt.figure(figsize=(6,4))
overtime_attrition_rate.plot(kind='bar')
plt.title("Attrition Rate (%) by Overtime")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

Employees working overtime show significantly higher attrition rates compared to those without overtime.

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[17, 25, 35, 45, 55, 65],
    labels=['18-25', '26-35', '36-45', '46-55', '56+']
)
```

```python
age_attrition_rate = (
    df.groupby('AgeGroup', observed=True)['AttritionFlag']
    .mean()
) * 100
```

```python
plt.figure(figsize=(8,5))
age_attrition_rate.plot(kind='bar')
plt.title("Attrition Rate (%) by Age Group")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

Younger employees (18–25 and 26–35) have higher attrition, suggesting early-career instability.

```python
df['TenureGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins=[-1, 2, 5, 10, 20, 40],
    labels=['0-2', '3-5', '6-10', '11-20', '20+']
)
```

```python
tenure_attrition_rate = (
    df.groupby('TenureGroup', observed=True)['AttritionFlag']
    .mean()
) * 100
```

```python
plt.figure(figsize=(8,5))
tenure_attrition_rate.plot(kind='bar')
plt.title("Attrition Rate (%) by Tenure Group")
plt.ylabel("Attrition Rate (%)")
plt.show()
```

Attrition decreases as years at company increase, showing that long-term employees are more stable.

Business Recommendations:

1. Introduce salary revisions or performance-based incentives for lower salary bands.
2. Reduce overtime workload or compensate overtime employees fairly.
3. Implement retention programs for early-career employees.
4. Improve employee engagement in high-attrition departments.
5. Monitor job satisfaction levels regularly to prevent employee burnout.
